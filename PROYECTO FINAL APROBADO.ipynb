Adem√°s, si se supone que un operador debe realizar llamadas salientes, un n√∫mero reducido de ellas tambi√©n ser√° un signo de ineficacia.

# Descripci√≥n de los datos

Los datasets contienen informaci√≥n sobre el uso del servicio de telefon√≠a virtual CallMeMaybe. Sus clientes son organizaciones que necesitan distribuir gran cantidad de llamadas entrantes entre varios operadores, o realizar llamadas salientes a trav√©s de sus operadores. Los operadores tambi√©n pueden realizar llamadas internas para comunicarse entre ellos. Estas llamadas se realizan a trav√©s de la red de CallMeMaybe.

El dataset comprimido _**telecom_dataset_us.csv**_ contiene las siguientes columnas:

- **`user_id`**: ID de la cuenta del cliente.
- **`date`**: Fecha en la que se recuperaron las estad√≠sticas.
- **`direction`**: "Direcci√≥n" de llamada (<font color='red'>out</font> para saliente, <font color='red'>in</font> para entrante).
- **`internal`**: Si la llamada fue interna (entre los operadores de un cliente o clienta).
- **`operator_id`**: Identificador del operador.
- **`is_missed_call`**: Si fue una llamada perdida.
- **`calls_count`**: N√∫mero de llamadas.
- **`call_duration`**: Duraci√≥n de la llamada (sin incluir el tiempo de espera).
- **`total_call_duration`**: Duraci√≥n de la llamada (incluido el tiempo de espera).

El conjunto de datos _**telecom_clients_us.csv**_ tiene las siguientes columnas:

- **`user_id`**: ID de usuario.
- **`tariff_plan`**: Tarifa actual de la clientela.
- **`date_start`**: Fecha de registro de la clientela.

# 1.2 Carga de Datos

Para la carga de los datos, se crea un bloque _try_ / _except_ con validaci√≥n de carga exitosa.
```python
# Carga de datos con validaci√≥n y manejo de errores
try:
    # Carga dataset con llamadas telef√≥nicas
    df_llamadas = pd.read_csv('https://raw.githubusercontent.com/WyldeBlack/Proyecto-Final-Telecomunicaciones/refs/heads/main/telecom_dataset_us.csv', # Link de GitHub
        parse_dates=['date'] # Convertir columna 'date' a formato datetime
    
    )
    # Carga dataset con informaci√≥n de clientes
    df_clientes = pd.read_csv('https://raw.githubusercontent.com/WyldeBlack/Proyecto-Final-Telecomunicaciones/refs/heads/main/telecom_clients_us.csv', # Link de GitHub
        parse_dates=['date_start'] # Convertir la columna 'date_start' a datetime
    )

    # Verificar que los DataFrames no est√©n vac√≠os
    if df_llamadas.empty: # Mensaje de error
        raise ValueError("El dataset de llamadas se encuentra vac√≠o. Revisar el archivo fuente") 
    if df_clientes.empty: # Mensaje de error
        raise ValueError("El dataset de clientes se encuentra vac√≠o. Revisar el archivo fuente") 

    # Mostrar resumen de carga de datos exitosa
    print("Datos cargados exitosamente:\n")
    print(f" - Registros en llamadas: {len(df_llamadas)}") # Total registros dataset de llamadas
    print(f" - Registros en clientes: {len(df_clientes)}\n") # Total registros dataset de clientes       

    # Separador
    print() 
    
    # Muestra inicial de datos
    print("Muestra de datos de llamadas:\n")
    display(df_llamadas.sample(5))

    # Separador
    print()
    
    print("Muestra de datos de clientes:\n")
    display(df_clientes.sample(5))

# Manejo de errores espec√≠ficos
except FileNotFoundError as e: # Mensaje error por archivo no encontrado
    print(f"\n Error: Archivo no encontrado. Verifica Rutas:")
    print(f"  - Detalles: {str(e)}")

except pd.errors.EmptyDataError: # Mensaje error por archivo vac√≠o
    print(f"\n Error: Uno o m√°s archivos est√°n vac√≠os.")

except ValueError as e: # Mensaje error por validaci√≥n
    print(f"\n Error de validaci√≥n")
    print(f"  - {str(e)}")

except Exception as e: # Mensaje error por cualquier otro tipo no especificado
    print(f"\n Error inesperado:")
    print(f"  - Tipo: {type(e).__name__}")
    print(f"  - Detalles {str(e)}")
```

# Asimetr√≠a Extrema (Media vs Mediana)

```python
# Datos para el gr√°fico
metricas = df_principal[['call_duration', 'total_call_duration', 'wait_time']].agg(['mean', 'median']).T # Definir variable 'metricas' y transponer para mejor visualizaci√≥n
metricas.rename(index={'call_duration':'Duraci√≥n Llamada Efectiva', # Renombrar columnas para mayor claridad en la visualizaci√≥n
                       'total_call_duration':'Duraci√≥n Total Llamada',
                       'wait_time':'Tiempo de Espera'}, inplace=True) 

# Gr√°fico de barras comparativas
metricas.plot(kind='bar', color=(['salmon','lightblue']) , edgecolor='black', figsize=(15,6), width=0.8) # Configurar gr√°fico de barras

# Agregar valores de media y mediana
for i, (mean_val, median_val) in enumerate(zip(metricas['mean'], metricas['median'])): # Bucle for para obtener los valores de media y mediana, combinando en pares y a√±adiendo un √≠ndice a cada par
    plt.text(i-0.2, mean_val+0.2, f'{mean_val:.1f}', ha='center') # Configuraci√≥n de texto y posici√≥n de texto para la media (con 1 decimal)
    plt.text(i+0.2, median_val+0.2, f'{median_val:.1f}', ha='center') # Configuraci√≥n de texto y posici√≥n de texto para la mediana (con 1 decimal)

plt.title('Comparaci√≥n Media vs Mediana', fontsize=20) # Configuraci√≥n de t√≠tulo del gr√°fico
plt.ylabel('Minutos', fontsize=15) # Configuraci√≥n de etiqueta eje y
plt.xticks(rotation=0, fontsize=15) # Configuraci√≥n rotaci√≥n (0) tama√±o de texto de etiquetas eje x
plt.legend(['Media (Promedio)', 'Mediana'], edgecolor='black' , fontsize=12) # Configuraci√≥n recuadro con leyenda

# Resaltar diferencia entre media y mediana
plt.text(-0.38, plt.ylim()[1]*0.95, 'Gran diferencia ‚Üí Asimetr√≠a extrema', fontsize=12, color='darkred', bbox=dict(facecolor='white', alpha=0.8)) # Configurar cuadro con texto
plt.show() # Mostrar gr√°fico
```

El gr√°fico confirma la diferencia extrema existente entre la media (promedio) y la mediana, en cuanto a las duraciones de las llamadas. Para confirmar estad√≠sticamente esta diferencia, que resulta en un sesgo hacia la derecha con cola muy larga, provocada por los valores extremos, se calcular√° el Coeficiente de Asimetr√≠a de Pearson, cuya formula es la siguiente:

$$
CA{p} = \frac{3(\bar{x} - M{o}(X))}{s}
$$

$$
D√≥nde:
$$

$$
\bar{x} = Media
$$

$$
M{o}(X) = Moda
$$

$$
s = Desviaci√≥n Est√°ndar
$$

El resultado de esta ecuaci√≥n debe variar entre **-3** y **3**, siendo **-3** una **_asimetr√≠a negativa_**, **0** representando una **_distribuci√≥n sim√©trica_** y **3** una **_asimetr√≠a positiva_**. Si el resultado es mayor, es producto de valores extremadamente altos y at√≠picos. 

```python
# Calcular Coeficiente de Asimetr√≠a de Pearson 
coef_asimetria = df_principal['call_duration'].skew() # Calcular asimetr√≠a en la distribuci√≥n de los datos
print(f"\nCoeficiente de Asimetr√≠a de Pearson: {coef_asimetria:.2f}\n")
```

Se confirma la asimetr√≠a positiva (y extrema).

# 2.5 An√°lisis de Operadores

Agrupaci√≥n de datos por operadores (**`operator_id`**), y c√°lculo de m√©tricas clave para el an√°lisis.

# Preparaci√≥n de Datos por Operador

```python
# Crear nuevo Df "operadores" agrupando por 'operator_id' y calcular m√©tricas
operadores = df_analisis.groupby('operator_id').agg({
    'calls_count':'sum', # Total de llamadas
    'is_missed_call':['sum', 'mean'], # Total de llamadas perdidas y promedio
    'call_duration':'mean', # Promedio de duraci√≥n de llamada
    'total_call_duration':'mean', # Promedio de duraci√≥n total de llamada
    'wait_time':'mean', # Promedio de tiempo de espera
    'direction': lambda x: (x == 'in').sum(), # Calcular cantidad de llamadas entrantes
}).reset_index()

# Cambiar nombre de columnas del DF operadores
operadores.columns = ['operator_id', 'total_calls', 'missed_calls', 'missed_rate', 'avg_call_duration', 'avg_total_duration', 'avg_wait_time', 'inbound_calls']

# Agregar la columna 'outbound_calls' y calcular las llamadas salientes
operadores['outbound_calls'] = operadores['total_calls'] - operadores['inbound_calls']

# Vista previa operadores
print("Vista previa operadores:\n")
display(operadores.head()) # Primeras 5 filas
print() # Separador
display(operadores.tail()) # √öltimas 5 filas
```

# Definir M√©tricas de Ineficacia

Es por ello que se filtrar√° esta informaci√≥n, creando `operadores_final`, que contendr√° solo las m√©tricas esenciales para la clasificaci√≥n de operadores ineficaces, seg√∫n los criterios planteados al comienzo de este an√°lisis. 

# Columnas Esenciales a Incluir en `operadores_final`: 

- **IDENTIFICACI√ìN**
  - **`operator_id`**: Identificador √∫nico del operador. La columna ser√° renombrada, eliminando el prefijo **"operator_"**, quedando finalmente solo como **`id`**
- **M√âTRICAS PRINCIPALES DE INEFICACIA**
  - **`operator_inbound_missed_rate`**: Tasa de llamadas entrantes perdidas (tarjet principal). Ser√° renombrada como **`inbound_missed_rate`**.
  - **`operator_inbound_wait_time`**: Tiempo de espera promedio en llamadas entrantes (tarjet principal). Ser√° renombrada como **`inbound_wait_time`**.
  - **`operator_outbound_rate`**: Proporci√≥n de llamadas salientes (para ineficacia moderada). Ser√° renombrada como **`outbound_rate`**.
- **M√âTRICAS COMPLEMENTARIAS**
  - **`total_calls`**: Volumen total de llamadas, para dar contexto al an√°lisis.
  - **`inbound_calls`**: Cantidad de llamadas entrantes, para validar significancia estad√≠stica.
  - **`outbound_calls`**: Cantidad de llamadas salientes, para dar contexto al an√°lisis.

# Exclusi√≥n de M√©tricas Generales:

- **`missed_calls`**, **`missed_rate`**, **`avg_call_duration`**, **`avg_total_duration`**, **`avg_wait_time`**: Ser√°n excluidas debido a que el an√°lisis se centra en llamadas entrantes, y estas m√©tricas generales pueden diluir los patrones de ineficiencia. 

```python
# Crear "operadores_final"
operadores_final = operadores[['operator_id', # Mantener ID de operador
                               'operator_inbound_missed_rate', # Mantener tasa de llamadas entrantes perdidas
                               'operator_inbound_wait_time', # Mantener tiempo de espera por operador para llamadas entrantes
                               'operator_outbound_rate', # Mantener tasa de llamadas salientes por operador
                               'total_calls', # Mantener total de llamadas
                               'inbound_calls', # Mantener llamadas entrantes
                               'outbound_calls' # Mantener llamadas salientes
                              ]].rename(columns={'operator_inbound_missed_rate':'inbound_missed_rate', # Eliminar prefijo "operator_"
                                                                 'operator_inbound_wait_time':'inbound_wait_time', # Eliminar prefijo "operator_"
                                                                 'operator_outbound_rate':'outbound_rate'}) # Eliminar prefijo "operator_"

# Mostrar estructura de "operadores_final"
print("==" * 14) # L√≠neas ornamentales
print("Estructura Operadores Final")
print("==" * 14) # L√≠neas ornamentales

# Informaci√≥n b√°sica del DataFrame
print("Informaci√≥n t√©cnica:\n")
print(operadores_final.info())
```

# DEFINICI√ìN DE UMBRALES PARA CLASIFICACI√ìN

```python
# Definir diccionario con umbrales clave para clasificaci√≥n de operadores
umbrales = {
    'p75_missed': operadores_final['inbound_missed_rate'].quantile(0.75), # Percentil 75 de tasa de llamadas perdidas (ineficacia moderada)
    'p90_missed': operadores_final['inbound_missed_rate'].quantile(0.90), # Percentil 90 de tasa de llamadas perdidas (ineficacia alta)
    'p75_wait' : operadores_final['inbound_wait_time'].quantile(0.75), # Percentil 75 de tiempo de espera en llamadas entrantes (ineficacia moderada)
    'p90_wait': operadores_final['inbound_wait_time'].quantile(0.90), # Percentil 90 de tiempo de espera en llamadas entrantes (ineficacia alta)
    'median_outbound': operadores_final['outbound_rate'].median() # Mediana de tasa de llamadas salientes (identifica  operadores con bajo volumen)
}

# Crear columna 'nivel_ineficacia' con valor "Eficaz" por defecto para todos los operadores
operadores_final['nivel_ineficacia'] = 'Eficaz'

# Definir condiciones para operadores con alta ineficacia
cond_alta = (
    (operadores_final['inbound_missed_rate'] >= umbrales['p90_missed']) & # Tasa de llamadas perdidas igual o superior al percentil 90
    (operadores_final['inbound_wait_time'] >= umbrales['p90_wait']) # Promedio de tiempo de espera igual o superior al percentil 90
)
operadores_final.loc[cond_alta, 'nivel_ineficacia'] = 'Alta' # Aplicar calsificaci√≥n "Alta" para los operadores que cumplan con la condici√≥n

# Definir condiciones para operadores con ineficacia moderada
cond_moderada = (
    (operadores_final['inbound_missed_rate'] >= umbrales['p75_missed']) & # Tasa de llamadas perdidas igual o superior al percentil 75
    (operadores_final['inbound_wait_time'] >= umbrales['p75_wait']) & # Promedio de tiempo de espera igual o superior al percentil 75
    (operadores_final['outbound_rate'] < umbrales['median_outbound']) & # Tasa de llamadas salientes menor a la mediana
    (~operadores_final['operator_id'].isin(operadores_final[cond_alta]['operator_id'])) # Evitar que se encuentren clasificados como "Alta", verificando por 'operator_id'
)
operadores_final.loc[cond_moderada, 'nivel_ineficacia'] = 'Moderada' # Aplicar clasificaci√≥n "Moderada" para los operadores que cumplan con la condici√≥n

# Verificaci√≥n
print("Distribuci√≥n de ineficacia:")
print(operadores_final['nivel_ineficacia'].value_counts()) # Distribuci√≥n de operadores seg√∫n nivel de ineficacia
print() # Separador
print('Tasa de ineficacia:')
print(operadores_final['nivel_ineficacia'].value_counts(normalize=True)) # Tasa de ineficacia por operadores
```

Los 1093 operadores se distribuyen, seg√∫n su nivel de ineficacia, de la siguiente manera:

- **_874_** operadores calificados con un nivel de eficiencia **Eficaz** (80%).
- **_173_** operadarores calificados con un nivel de ineficacia **Moderada** (16%).
- **_46_** operadores calificados con un nivel de ineficacia **Alta** (4%).

<center><h1>3. PRUEBAS DE HIP√ìTESIS</h1></center>

En esta √∫ltima etapa del an√°lisis de operadores ineficaces para la compa√±√≠a CallMeMaybe, se pondr√°n a prueba las hip√≥tesis planteadas al comienzo del proyecto:

- **_"Los operadores con mayor tiempo de espera tienen una tasa significativamente m√°s alta de llamadas perdidas"_**.
  
- **_"Las llamadas internas tienen menor tasa de p√©rdida que las llamadas externas"_**.

Se realizar√° cada prueba de hip√≥tesis por separado, documentando todo el proceso. 

# 3.1 "Los operadores con mayor tiempo de espera tienen una tasa significativamente m√°s alta de llamadas perdidas".

- Hip√≥tesis Nula (H‚ÇÄ): **_"No existe relaci√≥n significativa entre el tiempo de espera y la tasa de llamadas perdidas"_**.

- Hip√≥tesis Alternativa (H‚ÇÅ): **_"Existe una relaci√≥n positiva entre el tiempo de espera y la tasa de llamadas perdidas"_**.

Para desarrollar la primera prueba de hip√≥tesis, se definir√° una funci√≥n para determinar si las variables que se utilizar√°n, es decir, el tiempo de espera promedio para llamadas entrantes (**`inbound_wait_time`**) y la tasa de llamadas entrantes perdidas (**`inbound_missed_rate`**), presentan una distribuci√≥n normal o no. Para determinar la normalidad de las variables utilizar√© el **_Test de Shapiro-Wilk_**. En el caso de que la distribuci√≥n de ambas variables sea normal, se utilizar√° el **_Test de Correlaci√≥n de Pearson_** (param√©trico). En el caso contrario (distribuci√≥n no normal), se utilizar√° el **_Test de Correlaci√≥n de Spearman_** (no param√©trico). 

**TEST DE NORMALIDAD**

```python
# Definir funci√≥n "analisis_correlacion"
def analisis_correlacion(df, var1, var2, alpha=0.05, decimales=2):
    """
    Realiza un an√°lisis de correlaci√≥n autom√°tico seleccionando Pearson o Spearman, dependiendo de la normalidad de los datos.
    
    Par√°metros:
    
    df: Dataframe con los datos
    var1, var2: Strings con nombres de las columnas a analizar
    alpha: Nivel de significancia para determinar normalidad (0.05)
    decimales: N√∫mero de decimales a redondear (2)

    Retorna:
    Diccionario con resultados formateados y recomendaci√≥n
    """
    # Pruebas de normalidad
    shapiro1 = stats.shapiro(df[var1]) # Test de Shapiro-Wilk para variable 1
    shapiro2 = stats.shapiro(df[var2]) # Test de Shapiro-Wilk para variable 2

    # Seleccionar Test
    if shapiro1.pvalue > alpha and shapiro2.pvalue > alpha:
        corr, p = stats.pearsonr(df[var1], df[var2]) # Valores estad√≠sticos test de Pearson
        test = 'Pearson'
        simbolo = 'r'
    else:
        corr, p = stats.spearmanr(df[var1], df[var2]) # valores estad√≠sticos test de Spearman
        test = 'Spearman'
        simbolo = 'p'

    # Interpretaci√≥n de la fuerza de correlaci√≥n
    magnitud = ('Despreciable' if abs(corr) < 0.1 else # Definir nivel correlaci√≥n "Despreciable" (valor absoluto)
               'D√©bil' if abs(corr) < 0.3 else # Definir nivel de correlaci√≥n "D√©bil" (valor absoluto)
               'Moderada' if abs(corr) < 0.5 else # Definir nivel de correlaci√≥n "Moderada" (valor absoluto)
               "Fuerte") # Definir nivel de correlaci√≥n "Fuerte"

    # Formatear resultados
    resultado = {'test': test, # Recomendaci√≥n del test a utilizar ('Pearson'-'Spearman')
                 'coeficiente': corr, # Coeficiente de correlaci√≥n 
                 'p-value': p, # Valor estad√≠stico p
                 'normalidad_var1': shapiro1.pvalue, #
                 'normalidad_var2': shapiro2.pvalue, #
                 'resultado_str': ( # Strings a mostrar en el resultado del an√°lisis
                     f"Test a Utilizar: {test} ({simbolo} = {corr:.{decimales}f}, p = {p:.{decimales}f})\n" # Recomendaci√≥n de test a utilizar y los valores estad√≠sticos obtenidos 
                     f"Fuerza de Correlaci√≥n: {magnitud}\n" # Nivel de correlaci√≥n obtenido
                     f"Normalidad {var1}: W = {shapiro1.statistic:.{decimales}f}, p = {shapiro1.pvalue:.{decimales}f}\n" # Resultados estad√≠sticos para variable 1
                     f"Normalidad {var2}: W = {shapiro2.statistic:.{decimales}f}, p = {shapiro2.pvalue:.{decimales}f}"), # Resultados estad√≠sticos para variable 2
                 'interpretacion': (
                         "Los datos cumplen supuestos de normalidad" if test == 'Pearson' # Condici√≥n para imprimir interpretaci√≥n del an√°lisis 
                         else "Los datos NO cumplen supuestos de normalidad")}
    return resultado

# Aplicar la funci√≥n, y almacenar resultado en variable "resultados"
resultados = analisis_correlacion(operadores_final, 'inbound_wait_time', 'inbound_missed_rate')

# Imprimir resultados
print('==' * 19) # L√≠neas ornamentales
print('Resultados del An√°lisis de Normalidad')
print('==' * 19) # L√≠neas ornamentales
print() # Separador
print(resultados['resultado_str']) # Slice del resultado a imprimir ('resultado_str')
print('\nInterpretaci√≥n de Resultados:', resultados['interpretacion'] ) # Slice del resultado a imprimir ('interpretacion')
```







El resultado no indica que las variables **_NO cumplen con supuestos de normalidad_**, por lo que la prueba de hip√≥tesis se realizar√° utilizando el **_Test de Correlaci√≥n de Spearman_**.

**PRUEBA DE HIP√ìTESIS N¬∞1**

```python
# Definir funci√≥n "prueba_hipotesis_1"
def prueba_hipotesis_1(df, var1, var2, alpha=0.05, decimales=4):
    """
    Realiza prueba de hip√≥tesis de correlaci√≥n utilizando Spearman, prueba no param√©trica,
    incluye interpretaci√≥n completa.

    Par√°metros:

    df: Dataframe con los datos
    var1, var2: Strings con nombres de las columnas a analizar
    alpha: Nivel de significancia estad√≠stica (0.05)
    decimales: N√∫mero de decimales a redondear (4)

    Retorna:
    Diccionario con resultados estad√≠sticos y su interpretaci√≥n
    """
    # Calcular correlaci√≥n de Spearman
    corr, p_value = stats.spearmanr(df[var1], df[var2])

    # Determinar significancia estad√≠stica
    significativo = p_value < alpha

    # Interpretar fuerza de correlaci√≥n
    magnitud = ('Despreciable' if abs(corr) < 0.1 else # Definir nivel correlaci√≥n "Despreciable" (valor absoluto)
               'D√©bil' if abs(corr) < 0.3 else # Definir nivel correlaci√≥n "D√©bil" (valor absoluto)
               "Moderada" if abs(corr) <0.5 else # Definir nivel correlaci√≥n "Moderada" (valor absoluto)
               'Fuerte') # Definir nivel correlaci√≥n "Fuerte"

    # Formatear resultados
    resultado = {'correlacion': round(corr, decimales),
                'p_value': round(corr, decimales),
                'significativo': significativo,
                'interpretacion': (
                    f"Existe una correlaci√≥n {magnitud} y estad√≠sticamente {'significativa' if significativo else 'no significativa'}"
                    f"(p = {corr:.{decimales}f}, p = {p_value:.{decimales}f})"),
                'decision': (
                    'Se rechaza la hip√≥tesis nula: Existe evidencia de correlaci√≥n significativa' if significativo else
                    'No se puede rechazar la hip√≥tesis nula: No existe evidencia suficiente de correlaci√≥n')}

    return resultado

# Aplicar la funci√≥n, y almacenar resultado en variable "resultados"
resultados = prueba_hipotesis_1(operadores_final, 'inbound_wait_time', 'inbound_missed_rate')

# Imprimir resultados organizados
print('==' * 18)
print("RESULTADOS PRUEBA DE HIP√ìTESIS N¬∞1")
print('==' * 18)
print(f"Coeficiente de correlaci√≥n (œÅ): {resultados['correlacion']}")
print(f"Valor p: {resultados['p_value']}")
print(f"Significativo al nivel Œ±=0.05: {'S√≠' if resultados['significativo'] else 'No'}")
print("\nINTERPRETACI√ìN:")
print(resultados['interpretacion'])
print(resultados['decision'])
```

```python
# Configurar tama√±o del gr√°fico
plt.figure(figsize=(12, 6))

# Configurar gr√°fico
sns.regplot(data=operadores_final, x='inbound_wait_time', y='inbound_missed_rate', scatter_kws={'alpha': 0.5, 'color': 'blue'}, # Regresi√≥n lineal de operadores
    line_kws={'color': 'red', 'linestyle': '--'}) # Configurar recta de regresi√≥n 
plt.title(f'Correlaci√≥n Spearman: œÅ = {resultados["correlacion"]:.2f}, p = {resultados["p_value"]:.4f}') # Configurar t√≠tulo del gr√°fico (mostrando resultados obtenidos)
plt.xlabel('Tiempo de espera (minutos)') # Configurar etiqueta eje X
plt.ylabel('Tasa de llamadas perdidas') # Configurar etiqueta eje Y
plt.show() # Mostrar gr√°fico
```

**Hip√≥tesis nula rechazada**. Se confirma que existe evidencia de correlaci√≥n estad√≠sticamente significativa entre el tiempo de espera y la tasa de llamadas perdidas.

- Hip√≥tesis Nula (H‚ÇÄ): **_No hay diferencia en la tasa de p√©rdida entre llamadas internas y externas_**

- Hip√≥tesis Alternativa (H‚ÇÅ): **_Las llamadas internas tienen menor tasa de p√©rdida que las externas_**


Para desarrollar esta prueba de hip√≥tesis, revisar√© el promedio en las llamadas perdidas internas, para visualizar antes de realizar la prueba. El m√©todo aplicado ser√° el **_Test U de Mann-Whitney_**.

**PRUEBA DE HIP√ìTESIS N¬∞2**

```python
# Crear DF con llamadas internas perdidas
internas = df_analisis[df_analisis['internal'] == 1]['is_missed_call']

# Crear DF con llamadas externas perdidas
externas = df_analisis[df_analisis['internal'] == 0]['is_missed_call']

# Prueba unilateral (alternative='less': internas < externas)
stat, p = mannwhitneyu(internas, externas, alternative='less') 

print(f"Resultado MWU: U = {stat:.1f}, p = {p:.6f}") # f-string con valores obtenidos en la prueba
print("Rechazamos Hip√≥tesis Nula a favor de Hip√≥tesis Alternativa" if p < 0.05 else "No hay evidencia suficiente para rechazar Hip√≥tesis Nula") # String para rechazar o no la hip√≥tesis nula
```

```python
# Presentaci√≥n informe final con resultados
print('==' * 18) # L√≠neas ornamentales
print("RESULTADOS PRUEBA DE HIP√ìTESIS N¬∞2")
print('==' * 18) # L√≠neas ornamentales
print(f"\n- Muestra: {len(internas)} llamadas internas vs {len(externas)} externas") # f-string con el total de registros en llamadas perdidas externas e internas
print(f"- Tasa p√©rdida interna: {internas.mean():.1%} vs externa: {externas.mean():.1%}") # f-string con promedio de llamadas perdidas
print(f"- Prueba MWU unilateral: U = {stat:.1f}, p = {p:.6f}") # f-string con valores obtenidos en la prueba
print("\nConclusi√≥n:", "Las llamadas internas tienen significativamente menor tasa de p√©rdida" if p < 0.05 else "No hay diferencia significativa") # String para rechazar o no la hip√≥tesis nula
```
